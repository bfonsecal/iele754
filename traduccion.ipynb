```python
import numpy as np
import matplotlib.pyplot as plt
#DOG USUALLY BENIGN
# Dog eats random homework
# aka missing completely at random

N = 100
S = np.random.normal(size=N)
H = np.random.normal(0.5*S, size=N)
D = np.random.binomial(1, 0.5, size=N)
Hstar = H.copy()
Hstar[D == 1] = np.nan

plt.scatter(S, H, c='gray', edgecolors='none', linewidths=2)
plt.scatter(S, Hstar, c='red', s=60, edgecolors='none')
plt.xlabel('S')
plt.ylabel('H')
plt.show()
```

```python
#DOG PATH CAN BE BENIGN


N = 100
S = np.random.normal(size=N)
H = np.random.normal(0.5*S, size=N)

# dog eats 80% of homework where
D = np.random.binomial(1, np.where(S > 0, 0.8, 0), size=N)
Hstar = H.copy()
Hstar[D == 1] = np.nan

plt.scatter(S, H, c='gray', edgecolors='none', linewidths=2)
plt.scatter(S, Hstar, c='red', s=60, edgecolors='none')
plt.xlabel('S')
plt.ylabel('H')
plt.show()
```
